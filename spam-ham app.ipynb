```python
%%writefile spamHam.py
# %%writefile - Command to make .py file. webApp is name of the file.
# Streamlit is an open source app framework in Python language. It helps us create web apps for data science and machine learning in a short time.
import streamlit as st
import joblib
model= joblib.load('spam-ham')
st.title("What's your mail?")
ip= st.text_input('Enter your message here')
op= model.predict([ip])
if st.button('Predict'):
  st.title(op[0])       # Command to show output's first item, only executed when 'predict' button is pressed.
```
